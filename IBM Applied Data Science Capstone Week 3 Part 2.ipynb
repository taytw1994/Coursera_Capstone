```python
import pandas as pd

# could not get geocoder workihg, pulled data from csv file instead
url='https://cocl.us/Geospatial_data/Geospatial_Coordinates.csv'
df_geo=pd.read_csv(url)
display(df_geo.head())

#Part 1 CSV was saved with an unnamed index, causing a new column "Unnamed:0" to appear. Fixed by indicating index_col=[0]
df= pd.read_csv('Week3_Capstone_part1.csv',index_col=[0])
display(df.head())

df_merged = pd.merge(df,df_geo, on = 'Postal Code')
df_merged.head(11)

print('the merged dataset has',df_merged.shape[0], 'rows and', df_merged.shape[1], 'columns')

#Save csv for use in part 3 

df_merged.to_csv('Week3_Capstone_part2.csv',index=False)


```
